```python
import matplotlib.pyplot as plt
months=["jan",'feb','march','april','may','june','july']
sales=[120,150,170,180,160,150,180]
plt.plot(months,sales,linestyle=":",color="orange",marker='p',linewidth=3)
#line styles are (-, --, -., :)
plt.title("monthly sales")
plt.xlabel("months")
plt.ylabel("sales")
plt.grid()
plt.show()
```

```python
months=['jan','feb','march','april','may','june']
product_a=[20,25,30,35,40,77]
product_b=[15,22,28,38,45,70]
plt.plot(months,product_a,label="product_a")
plt.plot(months,product_b,label="product_b")
plt.xlabel("months")
plt.ylabel("sales")
plt.title("trends of both products")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

hours=[1,2,3,4,5,6,7,8]
marks=[35,37,40,55,56,58,70,75]
plt.title("hours studied vs marks")
plt.xlabel("hours")
plt.ylabel("marks")
plt.scatter(hours,marks,color='red',s=20)
plt.show()
```

```python
import matplotlib.pyplot as plt

hours=[1,2,3,4,5,6,7,8]
marks=[35,37,40,55,56,58,70,75]
colors=['red','blue','green','orange','black','purple','skyblue','pink']
plt.title("hours studied vs marks")
plt.xlabel("hours")
plt.ylabel("marks")
plt.scatter(hours,marks,c=colors,s=100)
plt.show()
```
