```python
data={
    "Marks":[35,77,56,54,98,70,77]
}
df=pd.DataFrame(data)
df
```

```python
def grades(mark):
  if mark>=50:
    return 'A'
  elif mark>=75:
    return 'B'
  elif mark>=60:
    return 'C'
  else:
    return 'D'
df['Grade']=(df['Marks'].apply(grades))
```

```python
data={
    "Department":[
        "CSE",
        "ECE",
        "EEE",
        "MECH",
        "CIVIL"
    ]
}

df=pd.DataFrame(data)
df
```

```python
le=LabelEncoder()
df["Department"]=le.fit_transform(df["Department"])
df
```

```python
print(
    dict(
        zip(
            le.classes_,
            le.transform(le.classes_)
        )
    )
)
```
