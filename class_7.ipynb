```python
#MIN-MAX SCALING
```

```python
# we will shrink all values between 0 and 1
```

```python
# x`=((x-xmin)/(xmax-xmin))
```

```python
data={
    "orderDate":[
        "2023-01-01",
        "2023-02-01",
        "2023-03-01",
        "2023-04-01",
    ],
    "sales":[
        1000,
        2000,
        3000,
        4900
    ]
}
df=pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
#here order date is a object(text)
```

```python
df['orderDate']=pd.to_datetime(df['orderDate'])
df.dtypes
```

```python
#the reason to convert it to extract useful information
```

```python
df["Year"]=(df['orderDate'].dt.year)
df
```

```python
df['Month']=(df['orderDate'].dt.month)
df
```

```python
df["monthName"]=(df['orderDate'].dt.month_name())
df
```

```python
df['Day']=(df["orderDate"].dt.day)
df
```

```python
df['dayname']=(df['orderDate'].dt.day_name())
df
```

```python
df['weekend']=(df["dayname"].isin(["Saturday","Sunday"]))

df
```

```python


today=pd.Timestamp.today()

df["DaysSinceOrder"]=(today-df["orderDate"]).dt.days
df
```

```python
df.sort_values("orderDate")
```
