```python
df['Active Cases']=(
    df["Confirmed"]
    -df['Deaths']
    -df["Recovered"]
)

print(df)
```

```python
df['Death Percentage']=(
    df["Deaths"]/df["Confirmed"]
)*100

print(df)
```

```python
Country_report=df.groupby(
    "Country/Region"
)[
    "Death Percentage"
].sum()
```

```python
print(Country_report.sort_values("Death Percentage",ascending=False).head())
```
