```python
from scipy.stats import ttest_1samp
salary=[

        50000,52000,51000,53000,55000,54000,56000,55000,58000
 ]
```

```python
#test claim average salary=45000
```

```python
#p-value decision and conclusion
```

```python
np.mean(salary)
#null hypothesis avg salary=45000
#alt avg salary!=45000
```

```python
t_stat,p=ttest_1samp(salary,45000)

print("p-value",p)
```

```python
#there is sufficient evidence to conclude that the avg salary is not 45000
```

```python
# the sample strategy suggests that the actual average salary is different from 45000
```

```python
import numpy as np
from scipy import stats
data=[
    45000,50000,52000,48000,51000,53000,55000,49000,52000
]

confidence=0.95

interval=stats.t.interval(
    confidence,
    len(data)-1,
    loc=np.mean(data),
    scale=stats.sem(data)
)

print(interval)
```

```python
import numpy as np
from scipy import stats
salary=np.random.randint(
    30000,100000,100
)
```

```python
mean_salary=np.mean(salary)
```

```python
interval=stats.t.interval(
    0.95,
    len(salary)-1,
    loc=mean_salary,
    scale=stats.sem(salary)
)
```

```python
print("Average_salary",mean_salary)

print("95% confidence interval",interval)
```
