```python
highest_marks=np.max(marks)
highest_index=np.argmax(marks)
topper_name=student_names[highest_index]
print("highest marks:",highest_marks)
print("topper name:",topper_name)
```

argmax will give you the index position of maximum value

find out the person name with highest attendance

```python
highest_att=np.max(attendance)
highest_index=np.argmax(attendance)
high_attendance=student_names[highest_index]
print("highest attendance:",highest_att)
print("topper name:",high_attendance)
```

find out who got the lowest marks and attendance

```python
lowest_att=np.min(attendance)
lowest_index=np.argmin(attendance)
low_attendance=student_names[lowest_index]
print("lowest attendance:",lowest_att)
print("low attendance name:",low_attendance)
```

```python
import numpy as np 
house_names=np.array([
    "house A",
    "house B",
    "house C",
    "house D",
    "house E",
    "house F",
    "house G",
    "house H"
])
```

```python
electricity_units=np.array([
    120,125,180,122,290,300,400])
```

```python
print("___smart eletricity analysis system______")
print("\n house names")
print(house_names)
```

```python
print("_____basic analytics___")
print("average consumption",np.mean(electricity_units))
print("maximum consumption",np.max(electricity_units))
print("minimum consumption",np.min(electricity_units))
print("total electricity consumption",np.sum(electricity_units))
print("\n highest consumption house")
highest_index=np.argmax(electricity_units)
highest_house=house_names[highest_index]
highest_units=electricity_units[highest_index]
print("house name",highest_house)
print("units consumed",highest_units)
```
